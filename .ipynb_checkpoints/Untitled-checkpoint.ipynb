```python
#1) Вычислите общую выручку за июль 2021 по тем сделкам, приход денежных средств которых не просрочен.

import pandas as pd

schedules = pd.read_excel('data.xlsx')

start_month = 'Июль 2021'
end_month = 'Август 2021'

# Находим индексы начального и конечного месяца
start_index = schedules[schedules['status'] == start_month].index.min()
end_index = schedules[schedules['status'] == end_month].index.max()

filtered_data = schedules.iloc[start_index + 1 : end_index]

status = "ПРОСРОЧЕНО"
july_2021_data = filtered_data.query('status != @status')

res = july_2021_data['sum'].sum()

print('Сумма = ', res)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

schedules = pd.read_excel('data.xlsx')
schedules['receiving_date'] = pd.to_datetime(schedules['receiving_date'], format='%d.%m.%y', errors='coerce')

start_month = 'Июль 2021'
end_month = 'Август 2021'

# Находим индексы начального и конечного месяца
start_index = schedules[schedules['status'] == start_month].index.min()
end_index = schedules[schedules['status'] == end_month].index.max()

filtered_data = schedules.iloc[start_index + 1 : end_index]

status = "ПРОСРОЧЕНО"
july_2021_data = filtered_data.query('status != @status')

daily_revenue = july_2021_data.groupby(july_2021_data['receiving_date'])['sum'].sum()

plt.figure(figsize=(10, 5))
daily_revenue.plot(kind='line', marker='o', color='skyblue')
plt.title('Выручка компании по дням за июль 2021 года')
plt.xlabel('Дата')
plt.ylabel('Выручка')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()

```

```python
# 3)Кто из менеджеров привлек для компании больше всего денежных средств в сентябре 2021?

import pandas as pd

schedules = pd.read_excel('data.xlsx')
schedules['receiving_date'] = pd.to_datetime(schedules['receiving_date'], format='%d.%m.%y', errors='coerce')

start_month = 'Сентябрь 2021'
end_month = 'Октябрь 2021'

# Находим индексы начального и конечного месяца
start_index = schedules[schedules['status'] == start_month].index.min()
end_index = schedules[schedules['status'] == end_month].index.max()

filtered_data = schedules.iloc[start_index + 1 : end_index]

managers =  filtered_data.groupby('sale')['sum'].sum()

top_manager = managers.idxmax()

print(top_manager, "из менеджеров привлек(ла) для компании больше всего денежных средств в сентябре 2021")




```

```python
# 4)  Какой тип сделок (новая/текущая) был преобладающим в октябре 2021?
import pandas as pd

schedules = pd.read_excel('data.xlsx')
schedules['receiving_date'] = pd.to_datetime(schedules['receiving_date'], format='%d.%m.%y', errors='coerce')

start_month = 'Октябрь 2021'
#end_month = 'Октябрь 2021'

# Находим индексы начального и конечного месяца
start_index = schedules[schedules['status'] == start_month].index.min()
# end_index = schedules[schedules['status'] == end_month].index.max()

filtered_data = schedules.iloc[start_index + 1 : ]

new =  filtered_data[filtered_data['new/current'] == 'новая']['client_id'].count()

current = filtered_data[filtered_data['new/current'] == 'текущая']['client_id'].count()


print(f"{'новая' if new > current else 'текущая'} тип сделки был преобладающим в октябре 2021")

```

```python
# 5) Сколько оригиналов договора по майским сделкам было получено в июне 2021?
import pandas as pd


schedules = pd.read_excel('data.xlsx')
schedules['receiving_date'] = pd.to_datetime(schedules['receiving_date'], format='%d.%m.%y', errors='coerce')

start_month = 'Июнь 2021'
end_month = 'Июль 2021'

# Находим индексы начального и конечного месяца
start_index = schedules[schedules['status'] == start_month].index.min()
end_index = schedules[schedules['status'] == end_month].index.max()

filtered_data = schedules.iloc[start_index + 1 : end_index]

may_2021_receiving = filtered_data[filtered_data['receiving_date'].dt.month == 5]

original_count = may_2021_receiving[may_2021_receiving['document'] == 'оригинал']['client_id'].count()


print(f"{original_count} - столько оригиналов договора по майским сделкам было получено в июне 2021")
```

```python
# Не очень понятны данные, вот так выглядит примерная логика

import pandas as pd

schedules = pd.read_excel('data.xlsx')


schedules['receiving_date'] = pd.to_datetime(schedules['receiving_date'], format='%d.%m.%y', errors='coerce')


cutoff_date = pd.to_datetime('2021-07-01')


filtered_deals = schedules[schedules['receiving_date'] < cutoff_date]

# Вычисление бонусов для новых сделок
new_deals_bonus = filtered_deals[
    (filtered_deals['new/current'] == 'новая') &
    (filtered_deals['status'] == 'ОПЛАЧЕНО') &
    (filtered_deals['document'] == 'оригинал')
].copy()

new_deals_bonus['bonus'] = new_deals_bonus['sum'] * 0.07  # 7% бонуса

# Вычисление бонусов для текущих сделок
current_deals_bonus = filtered_deals[
    (filtered_deals['new/current'] == 'текущая') &
    (filtered_deals['document'] == 'оригинал') &
    (filtered_deals['status'] != 'ПРОСРОЧЕНО')
].copy()

current_deals_bonus['bonus'] = current_deals_bonus['sum'].apply(
    lambda x: x * 0.05 if x > 10000 else x * 0.03  # 5% или 3% бонуса
)

# Объединение бонусов по новым и текущим сделкам
all_bonuses = pd.concat([new_deals_bonus[['sale', 'bonus']], current_deals_bonus[['sale', 'bonus']]])

# Учет остатков для оригиналов, полученных позже 01.07.2021
# Фильтруем по оригиналам, полученным позже 01.07.2021
remaining_deals = filtered_deals[filtered_deals['receiving_date'] > cutoff_date]

# Фильтруем по менеджерам, у которых есть сделки, не имеющие оригиналов
remaining_bonuses = all_bonuses[all_bonuses['sale'].isin(remaining_deals['sale'])]

# Группировка по менеджерам и суммирование бонусов
final_bonus = remaining_bonuses.groupby('sale')['bonus'].sum().reset_index()

# Вывод остатков бонусов для каждого менеджера
print(final_bonus)
```
